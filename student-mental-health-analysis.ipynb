# Introduction

Mental health is an important factor affecting student well-being and academic performance. Students face academic pressure, lifestyle imbalance, and personal challenges that may lead to stress, anxiety, and depression.

This project aims to analyze student mental health data to identify patterns and relationships between academic performance and mental health indicators.

# Objectives

The objectives of this analysis are:

- To understand the distribution of student mental health conditions
- To analyze the relationship between CGPA and depression
- To analyze gender-wise mental health distribution
- To identify key factors affecting student mental health

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully")
```

```python
df = pd.read_csv("/kaggle/input/student-mental-health/Student Mental health.csv")
df.head()
```

# Dataset Overview

This section shows the structure and basic information of the dataset.

```python
df.info()
```

```python
df.describe()
```

# Data Cleaning

This step checks for missing values and removes incomplete data.

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
print("Missing values removed")
```

# Gender Distribution

This graph shows the number of male and female students.

```python
sns.countplot(x='Choose your gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

# Depression Distribution

This graph shows how many students experience depression.

```python
sns.countplot(x='Do you have Depression?', data=df)
plt.title("Depression Distribution")
plt.show()
```

# Anxiety Analysis

This graph shows how many students experience anxiety.

```python
sns.countplot(x='Do you have Depression?', data=df)
plt.title("Anxiety Distribution")
plt.show()
```

# CGPA vs Depression Analysis

This graph shows the relationship between academic performance and depression.

```python
sns.countplot(x='Choose your gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

# Correlation Analysis

This heatmap shows relationships between numerical variables.

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
numeric_df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

# Key Findings

Based on the analysis:

- Some students experience depression and anxiety
- Academic performance shows relationship with mental health
- Mental health issues are present across both genders
- Mental health is influenced by multiple factors

# Conclusion

This project analyzed student mental health using data analysis techniques. The results show that mental health issues such as anxiety and depression exist among students and may be influenced by academic performance and other factors.

This analysis helps in understanding student mental health and highlights the importance of mental health support systems.
